```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_nodum = train_df.drop('target', axis=1)
X = pd.get_dummies(X_nodum)
y = train_df['target']
c = {'low_risk' : 0, 'high_risk' : 1}   
y = y.replace(c)
```

```python
X.info()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_nodum = test_df.drop('target', axis=1)
X_test = pd.get_dummies(X_nodum)
y_test = test_df['target']
c = {'low_risk' : 0, 'high_risk' : 1}   
y_test = y_test.replace(c)
X_test.info()
```

```python
# add missing dummy variables to testing set
X_test['debt_settlement_flag_Y'] = 0
```

### Performance expectations

Dataset has many variables. It may be difficult for a logistic regression model to produce satisfactory parameters to account for all possibilities if it cannot determine which features a deterministic and which are not.
A random forests model may have more success in categorising parameters.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X, y)
print(f"Score with training data: {model.score(X, y)}")
print(f"Score with test data: {model.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, y)
print(f"Score with training data: {model.score(X, y)}")
print(f"Score with test data: {model.score(X_test, y_test)}")
```

### Performance realities

Neither model performed well with the test data, although the random forest model performed slightly better than the linear regression model.  
The failure of the Logistic Regression to converge, despite the high number of iterations allowed, is consistent with the expectation that this model would struggle to successfully identify parameters given the large number of features. Testing showed that increasing the number of iterations had a slight improvement in model score, but not significantly.  
The random forest model performed very well with the training data, suggesting that the test data does not behave in the same way as the training data for some reason

In both cases, the models are not reliable enough to be used on real data

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
X_train_scaled = scaler.transform(X)
X_test_scaled = scaler.transform(X_test)
```

### Revised expectations

Having scaled the data, it would be reasonable to expect performance to improve. The same reasoning as above would expect the random forest model to perform better than the linear regression model.

```python
# Train the Logistic Regression model on the scaled data and print the model score
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y)
print(f"Score with training data: {model.score(X_train_scaled, y)}")
print(f"Score with test data: {model.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
model = RandomForestClassifier()
model.fit(X_train_scaled, y)
print(f"Score with training data: {model.score(X_train_scaled, y)}")
print(f"Score with test data: {model.score(X_test_scaled, y_test)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train_scaled, y)
print(f"Score with training data: {model.score(X_train_scaled, y)}")
print(f"Score with test data: {model.score(X_test_scaled, y_test)}")
```

### Revised performance

As expected, the linear regression model performed better once the data was scaled. Importantly, it has converged on a solution, which it did not do earlier.  
The random forest classifier model has an unchanged performance, highlighting that scaling data makes no difference to this type of model.
Using a random forest regressor model instead, taking advantage of the scaled data, shows no improvement in performance; indeed the model performs worse than a random predictor.

This underlines that there are differences in credit risk between the two years which are not explained by the training dataset (2019)
